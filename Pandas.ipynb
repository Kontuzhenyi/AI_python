```python
# Импортируем библиотеки
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Читаем датасет
# ссылка https://archive.ics.uci.edu/dataset/748/sirtuin6+small+molecules-1
data = pd.read_csv('SIRTUIN6.csv', sep=',', encoding='cp1252')
print(data.head(10))
```

```python
# Выводим информацию
data.info()
```

```python
# Выводим информацию о типах столбцов
data.dtypes
```

```python
# Названия столбцов и строк
print('Столбцы: ', list(data.columns.values))
print('Строки: ', list(data.index.values))
```

```python
# Количество пустых ячеек
print(np.count_nonzero(data.isnull().sum(axis= 1).values))
```

```python
# Заменяем категориальные данные
data['Class'], _ = pd.factorize(data['Class'])
print(data.head(10))
```

```python
# Выводим информацию
data.info()
```

```python
# Визуализация
plt.hist(data['Class'])
plt.xlabel('Высокий или низкий уровень BFE')
plt.ylabel('Количество')
plt.show()
# Их примерно поровну
```

```python
# Матрица корреляции
plt.figure(figsize=(15, 10)).add_subplot(1,1,1)
sns.heatmap(data.corr(), annot = True)
plt.show()
```

```python
plt.hist(data['SC-5'])
plt.title('Распределение SC-5')
plt.show()
# В дипозоне от 0.3 до 0.4 наибольшее количество
```

```python
plt.hist(data['SP-6'])
plt.title('Распределение SP-6')
plt.show()
# Сложно что-то сказать
```

```python
plt.hist(data['SHBd'])
plt.title('Распределение SHBd')
plt.show()
# Сложно что-то сказать
```

```python
data.plot(kind='scatter', x='Class', y='FMF')
plt.show()
# Тип BFE не зависит от FMF
```

```python
# Зависимости всех столбцов
sns.pairplot(data, diag_kind='kde')
plt.show()
# Возможные взаимосвязи между дескрипторами, а также с BFE
```

```python
# Диаграмма рассеяния с линейной регрессией
for column in data.columns:
    if column != 'Class':
        sns.lmplot(x=column, y='Class', data=data, aspect=2, height=4)
        plt.title(f'Regression of BFE vs {column}')
        plt.show()
```
